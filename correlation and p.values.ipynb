```python
df = pd.read_csv(r'C:\Users\PC\Downloads\archive (3)\insurance.csv')
df_encoded = pd.get_dummies(df)
df_encoded.corr() #correlation is done only on numerical data this code import the csv and convert non numerical values
```

```python
df.charges.corr(df.bmi) #the correlation between bmi and charges
```

```python
from scipy import stats
corr, p_value = stats.pearsonr(df['charges'], df['age'])
print(round(corr,1))
print(round(p_value,29)) #returns the p value and correlation between charges and age and round off the decimals
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.age,df.charges)
plt.title('age vs charges')
plt.show() #visualizing to determine the correlation and p_value. Also naming the scatterplot
```
